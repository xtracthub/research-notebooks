```python
import matplotlib.pyplot as plt
import numpy as np
import itertools as IT
import math

# read the data into a dictionary object

minval = 1605223836
t_files = {} # dictionary with keys being times and values being num files processed at that time
#print(t_files[2])

raw_data = np.genfromtxt("crawl_data.csv", delimiter=",", names=["f_id", "crawl_timestamp", "num_files", "num_bytes"])
#print(raw_data["crawl_timestamp"])
for x, y in zip(raw_data["crawl_timestamp"], raw_data["num_files"]):
    t_files.setdefault(math.floor(x)-minval, 0)
    t_files[math.floor(x)-minval] += y

print("all done")
#print(t_files)
```

```python
# sort dictionary data and put it into two arrays

t_files_sorted = sorted(t_files.items())

sorted_timestamp = []
sorted_files = []

cumulative_files = 0
for x in t_files_sorted:
    sorted_timestamp.append(x[0])
    cumulative_files += x[1]
    sorted_files.append(cumulative_files)

print("all done")
#print(sorted_timestamp)
#print(sorted_files)
```

```python
# plot time vs files processed

plt.plot(sorted_timestamp, sorted_files)
plt.title('cumulative files processed over time')
plt.xlabel('time elapsed (seconds)')
plt.ylabel('number of files processed')
plt.show()
print("all done")
```

```python
# read the data into a dictionary object

t_bytes = {} # dictionary with keys being times and values being num bytes processed at that time

for x, y in zip(raw_data["crawl_timestamp"], raw_data["num_bytes"]):
    t_bytes.setdefault(math.floor(x)-minval, 0)
    t_bytes[math.floor(x)-minval] += y

print("all done")
#print(t_bytes)
```

```python
# sort dictionary data and put it into two arrays

t_bytes_sorted = sorted(t_bytes.items())

# use same sorted_timestamp array from files graph for bytes graph as well
sorted_bytes = []

cumulative_bytes = 0
for x in t_bytes_sorted:
    cumulative_bytes += x[1]
    sorted_bytes.append(cumulative_bytes)

print("all done")
#print(sorted_timestamp)
#print(sorted_bytes)
```

```python
# plot time vs bytes processed

plt.plot(sorted_timestamp, sorted_bytes)
plt.title('cumulative bytes processed over time')
plt.xlabel('time elapsed (seconds)')
plt.ylabel('number of bytes processed')
plt.show()
print("all done")
```
